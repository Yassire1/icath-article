### Step 2: Download C-MAPSS (NASA Turbofan)
**Priority: HIGH** - This is the primary RUL benchmark dataset

```python
def download_file(url, dest, chunk_size=8192):
    """Download file with progress bar"""
    response = requests.get(url, stream=True)
    response.raise_for_status()
    total_size = int(response.headers.get('content-length', 0))
    
    with open(dest, 'wb') as f:
        with tqdm(total=total_size, unit='iB', unit_scale=True, desc=dest.name) as pbar:
            for chunk in response.iter_content(chunk_size=chunk_size):
                size = f.write(chunk)
                pbar.update(size)
    return True

# C-MAPSS download
cmapss_dir = RAW_DIR / "cmapss"
cmapss_dir.mkdir(parents=True, exist_ok=True)

print("C-MAPSS Dataset (NASA Turbofan)")
print("=" * 60)
print("Option 1: Direct download from NASA")
print("URL: https://data.nasa.gov/download/ff5v-kuh6/application%2Fzip")
print()
print("Option 2: Manual download")
print("1. Go to: https://data.nasa.gov/Aerospace/CMAPSS-Jet-Engine-Simulated-Data/ff5v-kuh6")
print("2. Download CMAPSSData.zip")
print(f"3. Extract to: {cmapss_dir}")
print("=" * 60)
```

```python
# Try automatic download
try:
    zip_path = cmapss_dir / "CMAPSSData.zip"
    if not zip_path.exists():
        print("Downloading C-MAPSS...")
        download_file(
            "https://data.nasa.gov/download/ff5v-kuh6/application%2Fzip",
            zip_path
        )
        
        # Extract
        with zipfile.ZipFile(zip_path, 'r') as zf:
            zf.extractall(cmapss_dir)
        print("C-MAPSS extracted successfully!")
    else:
        print("C-MAPSS already downloaded")
    
    # Verify
    expected_files = ["train_FD001.txt", "test_FD001.txt", "RUL_FD001.txt"]
    for f in expected_files:
        if (cmapss_dir / f).exists():
            print(f"  {f}: OK")
        else:
            print(f"  {f}: MISSING")
            
except Exception as e:
    print(f"Automatic download failed: {e}")
    print("Please download manually from the URL above")
```

### Step 6: Download MIMII Dataset (Automated)

```python
mimii_dir = RAW_DIR / "mimii"
mimii_dir.mkdir(parents=True, exist_ok=True)

mimii_urls = {
    "fan": "https://zenodo.org/record/3384388/files/fan.zip",
    "pump": "https://zenodo.org/record/3384388/files/pump.zip",
    "slider": "https://zenodo.org/record/3384388/files/slider.zip",
    "valve": "https://zenodo.org/record/3384388/files/valve.zip"
}

print("MIMII Dataset (Zenodo)")
print("=" * 60)

for name, url in mimii_urls.items():
    zip_path = mimii_dir / f"{name}.zip"
    if not zip_path.exists():
        print(f"Downloading {name}...")
        try:
            download_file(url, zip_path)
            with zipfile.ZipFile(zip_path, 'r') as zf:
                zf.extractall(mimii_dir / name)
            print(f"  {name}: OK")
        except Exception as e:
            print(f"  {name}: FAILED - {e}")
    else:
        print(f"  {name}: Already downloaded")

print("=" * 60)
```

### Step 8: Verify All Datasets

```python
print("=" * 60)
print("DATASET DOWNLOAD SUMMARY")
print("=" * 60)

datasets = {
    "C-MAPSS": {
        "dir": cmapss_dir,
        "check_files": ["train_FD001.txt"],
        "priority": "HIGH"
    },
    "PHM Milling": {
        "dir": phm_dir,
        "check_files": [],
        "priority": "MEDIUM"
    },
    "PU Bearings": {
        "dir": pu_dir,
        "check_files": [],
        "priority": "MEDIUM"
    },
    "Wind SCADA": {
        "dir": wind_dir,
        "check_files": [],
        "priority": "LOW"
    },
    "MIMII": {
        "dir": mimii_dir,
        "check_files": ["fan", "pump", "slider", "valve"],
        "priority": "MEDIUM"
    },
    "PRONOSTIA": {
        "dir": pronostia_dir,
        "check_files": [],
        "priority": "LOW"
    }
}

for name, info in datasets.items():
    status = "READY" if info["dir"].exists() else "MISSING"
    files_ok = all((info["dir"] / f).exists() for f in info["check_files"])
    if info["check_files"] and not files_ok:
        status = "INCOMPLETE"
    print(f"  {name} [{info['priority']}]: {status}")

print("=" * 60)
print("\nNext: Run Notebook 03 (Preprocessing)")
```
